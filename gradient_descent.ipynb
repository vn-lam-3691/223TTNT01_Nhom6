```python
import math
import numpy as np
import matplotlib.pyplot as plt

def grad(w):
    return w*2 + 5*np.cos(w)

def loss(w):
    return w**2 + 5*np.sin(w)

x_asis = np.arange(-5, 5)
y_asis = loss(x_asis)
plt.plot(x_asis, y_asis)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
def gradient_descent(alpha, w_init, iteration, tolerance):
    w_list = [w_init]
    for it in range(iteration):
        print('loss = {:.3f} at iteration {:d}'.format(loss(w_list[-1]), it))
        w_new = w_list[-1] - alpha*grad(w_list[-1])
        if abs(grad(w_new)) < tolerance:
            break
        w_list.append(w_new)
    return w_list, it

alpha = 0.1
w_init = 4
tolerance = 0.001
iteration = 100
w_list, iteration_number = gradient_descent(alpha, w_init, iteration, tolerance)

print('Minima at w = {:.3f}, loss = {:.3f}, obtained after {:d} iterations'.format(w_list[-1], loss(w_list[-1]), iteration_number))
```

```python
loss_list = loss(np.array(w_list))
plt.scatter(w_list, loss_list, c = w_list)
plt.plot(x_asis, y_asis)
plt.xlabel('X')
plt.ylabel('Y')
plt.figure(figsize=(5,5), dpi=300)
plt.show()
```

```python
alpha = 0.05
w_list, iteration_number = gradient_descent(alpha, w_init, iteration, tolerance)

loss_list = loss(np.array(w_list))
plt.scatter(w_list, loss_list, c = w_list)
plt.plot(x_asis, y_asis)
plt.xlabel('X')
plt.ylabel('Y')
plt.figure(figsize=(5,5), dpi=300)
plt.show()
```
