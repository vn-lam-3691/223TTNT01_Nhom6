```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs # tạo dữ liệu
from sklearn.cluster import KMeans # xây dựng thuật toán kmeans
```

```python
x, _ = make_blobs(n_samples=200, centers=3, random_state=99, cluster_std=2.75)  # tạo 3 cụm dữ liệu, xoay quanh 200 điểm ngẫu nhiên
```

```python
x
```

```python
x1 = x[:,0]  # lấy tất cả giá trị trong cột 0
x2 = x[:,1]  # lấy tất cả giá trị trong cột 0
plt.scatter(x1, x2)
```

```python
# Bắt đầu thuật toán k-means
kmeans = KMeans(n_clusters=3, random_state=99)
```

```python
clustering_output = kmeans.fit(x)
```

```python
clustering_output.labels_
```

```python
clustering_output.cluster_centers_
```

```python
plt.scatter(x1, x2, color='blue')
plt.scatter(clustering_output.cluster_centers_[:,0], clustering_output.cluster_centers_[:,1], color='red')
```

```python
colors_dict = {0: 'blue', 1: 'green', 2: 'yellow'}
colors = [colors_dict[i] for i in kmeans.labels_]
```

```python
plt.scatter(x1, x2, color=colors)
plt.scatter(clustering_output.cluster_centers_[:,0], clustering_output.cluster_centers_[:,1], color='red')
```
